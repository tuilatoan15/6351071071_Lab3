#Câu c

```python
print("Nguyễn Hữu Toàn – 6351071071")

# Đọc dữ liệu Iris
from sklearn.datasets import load_iris
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

iris = load_iris()
X = iris.data       # 150x4
y = iris.target     # 150 nhãn 0,1,2
target_names = iris.target_names

#Chuẩn hóa dữ liệu về [0,1]
X = X / X.max(axis=0)

# Định nghĩa hàm create_model
def create_model():
    model = Sequential()
    # Hidden layer 1
    model.add(Dense(10, activation='relu', input_shape=(X.shape[1],)))
    # Hidden layer 2
    model.add(Dense(20, activation='relu'))
    # Output layer
    model.add(Dense(len(np.unique(y)), activation='softmax'))

    # Compile
    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    return model

print("Data, model đã sẵn sàng!")
```

#Câu d

```python
print("Nguyễn Hữu Toàn – 6351071071")

from sklearn.model_selection import KFold
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, f1_score

# Thiết lập k-fold
k = 10
kf = KFold(n_splits=k, shuffle=True, random_state=42)

# Lưu kết quả
models = []
accuracy_per_fold = []
precision_per_fold = []
recall_per_fold = []
f1_per_fold = []

fold_no = 1
for train_index, val_index in kf.split(X):
    print(f"\n--- Fold {fold_no} ---")

    X_train_fold, X_val_fold = X[train_index], X[val_index]
    y_train_fold, y_val_fold = y[train_index], y[val_index]

    # Tạo model mới
    model = create_model()

    # Huấn luyện
    history = model.fit(X_train_fold, y_train_fold,
                        validation_data=(X_val_fold, y_val_fold),
                        epochs=100, batch_size=1, verbose=0)

    # Dự đoán validation fold
    y_val_pred_probs = model.predict(X_val_fold)
    y_val_pred = np.argmax(y_val_pred_probs, axis=1)

    # Tính các metrics
    acc = accuracy_score(y_val_fold, y_val_pred)
    prec = precision_score(y_val_fold, y_val_pred, average='macro')
    rec = recall_score(y_val_fold, y_val_pred, average='macro')
    f1 = f1_score(y_val_fold, y_val_pred, average='macro')

    print(f"Accuracy: {acc:.4f}, Precision: {prec:.4f}, Recall: {rec:.4f}, F1-score: {f1:.4f}")

    # Lưu kết quả
    accuracy_per_fold.append(acc)
    precision_per_fold.append(prec)
    recall_per_fold.append(rec)
    f1_per_fold.append(f1)
    models.append(model)

    fold_no += 1

print("\nK-fold training hoàn tất!")
```

#Câu e

```python
print("Nguyễn Hữu Toàn – 6351071071")
import numpy as np

# Tính trung bình
avg_accuracy = np.mean(accuracy_per_fold)
avg_precision = np.mean(precision_per_fold)
avg_recall = np.mean(recall_per_fold)
avg_f1 = np.mean(f1_per_fold)

print("\n--- Average Metrics sau 10-fold CV ---")
print(f"Average Accuracy : {avg_accuracy:.4f}")
print(f"Average Precision: {avg_precision:.4f}")
print(f"Average Recall   : {avg_recall:.4f}")
print(f"Average F1-score : {avg_f1:.4f}")
```

#Câu f

```python
print("Nguyễn Hữu Toàn – 6351071071")
from sklearn.metrics import precision_recall_fscore_support

# Lưu metrics từng fold cho mỗi class
precision_per_class_folds = []
recall_per_class_folds = []
f1_per_class_folds = []

for train_index, val_index in kf.split(X):
    X_train_fold, X_val_fold = X[train_index], X[val_index]
    y_train_fold, y_val_fold = y[train_index], y[val_index]

    model = create_model()
    model.fit(X_train_fold, y_train_fold, epochs=100, batch_size=1, verbose=0)

    y_val_pred_probs = model.predict(X_val_fold)
    y_val_pred = np.argmax(y_val_pred_probs, axis=1)

    prec, rec, f1, _ = precision_recall_fscore_support(y_val_fold, y_val_pred, average=None)
    precision_per_class_folds.append(prec)
    recall_per_class_folds.append(rec)
    f1_per_class_folds.append(f1)

# Chuyển sang numpy để tính trung bình
precision_per_class_folds = np.array(precision_per_class_folds)
recall_per_class_folds = np.array(recall_per_class_folds)
f1_per_class_folds = np.array(f1_per_class_folds)

avg_precision_per_class = np.mean(precision_per_class_folds, axis=0)
avg_recall_per_class = np.mean(recall_per_class_folds, axis=0)
avg_f1_per_class = np.mean(f1_per_class_folds, axis=0)

print("\n--- Average Metrics theo từng class ---")
for idx, class_name in enumerate(target_names):
    print(f"{class_name}: Precision={avg_precision_per_class[idx]:.4f}, Recall={avg_recall_per_class[idx]:.4f}, F1-score={avg_f1_per_class[idx]:.4f}")
```

#Câu g

```python
print("Nguyễn Hữu Toàn – 6351071071")

# 3 mẫu mới
X_new = np.array([
    [6.2, 2.9, 4.3, 1.3],
    [5.1, 3.5, 1.4, 0.2],
    [7.3, 2.8, 6.4, 2.1]
])

# Chuẩn hóa giống dữ liệu train
X_new = X_new / X.max(axis=0)

# Tìm index model có F1-score tốt nhất
best_index = np.argmax([np.mean(f1) for f1 in f1_per_class_folds])  # lấy fold có F1 cao nhất
best_model = models[best_index]

# Dự đoán
y_new_pred_probs = best_model.predict(X_new)
y_new_pred = np.argmax(y_new_pred_probs, axis=1)

# In nhãn dự đoán
print("\n--- Dự đoán 3 mẫu mới ---")
for i, pred in enumerate(y_new_pred):
    print(f"Mẫu {i+1}: {X_new[i]} -> Dự đoán: {target_names[pred]}")
```

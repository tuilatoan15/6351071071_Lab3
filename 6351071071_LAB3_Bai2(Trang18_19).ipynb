#Câu 6

```python
print("Nguyễn Hữu Toàn – 6351071071")

# Gom cụm Tuổi và Spending Score với Elbow Method và K-Means
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

# Chọn dữ liệu
X1 = df[['Age', 'Spending Score (1-100)']].values

# Tính Inertia (WSS) cho k từ 1 đến 10
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init='random', random_state=42)
    kmeans.fit(X1)
    inertia.append(kmeans.inertia_)

# Vẽ biểu đồ Elbow
plt.figure(figsize=(10,6))
plt.plot(range(1,11), inertia, 'o-', color='b')
plt.xlabel('Số lượng cụm (k)')
plt.ylabel('WSS / Inertia')
plt.title('Phương pháp khuỷu tay (Elbow Method)')
plt.show()

# Chọn k=4 (theo hình Elbow) và thực hiện K-Means
kmeans = KMeans(n_clusters=4, init='random', random_state=42)
y_kmeans = kmeans.fit_predict(X1)

# In nhãn và tâm cụm
print("Nhãn các mẫu:", y_kmeans)
print("Tâm cụm:", kmeans.cluster_centers_)
```

#Câu 7

```python
print("Nguyễn Hữu Toàn – 6351071071")

# Biểu diễn kết quả gom cụm
plt.figure(figsize=(12,7))
plt.clf()

# Biểu đồ scatter các mẫu theo nhãn cụm
plt.scatter(x=df['Age'], y=df['Spending Score (1-100)'],
            c=y_kmeans, s=200, cmap='viridis', alpha=0.6, label='Mẫu')

# Vẽ tâm cụm
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],
            s=300, c='red', alpha=0.5, marker='X', label='Tâm cụm')

plt.xlabel('Tuổi')
plt.ylabel('Điểm chi tiêu (1-100)')
plt.title('Kết quả gom cụm K-Means (k=4)')
plt.legend()
plt.show()
```

#Câu 8

```python
print("Nguyễn Hữu Toàn – 6351071071")

from sklearn import metrics

# Đánh giá kết quả gom cụm bằng Silhouette coefficient
sil_score = metrics.silhouette_score(X1, y_kmeans, metric='euclidean')
print("Hệ số Silhouette trung bình của gom cụm Tuổi & Spending Score:", sil_score)
```

#Câu 9

```python
print("Nguyễn Hữu Toàn – 6351071071")

import pandas as pd
from sklearn.cluster import KMeans
import plotly.graph_objs as go
import plotly.offline as py

# Đọc dữ liệu
data = pd.read_csv("Mall_Customers.csv")  # đường dẫn tới file CSV của em
print(data.head())  # kiểm tra dữ liệu

# Gom cụm 3D
X3 = data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].values

kmeans3 = KMeans(n_clusters=4, init='random', random_state=42)
labels3 = kmeans3.fit_predict(X3)

data['label3'] = labels3

# biểu đồ 3D
trace1 = go.Scatter3d(
    x = data['Age'],
    y = data['Spending Score (1-100)'],
    z = data['Annual Income (k$)'],
    mode = 'markers',
    marker = dict(
        color = data['label3'],
        size = 8,
        line = dict(color=data['label3'], width=1),
        opacity = 0.8
    )
)

layout = go.Layout(
    title='Kết quả gom cụm 3D: Tuổi, Thu nhập & Spending Score',
    scene = dict(
        xaxis=dict(title='Tuổi'),
        yaxis=dict(title='Spending Score'),
        zaxis=dict(title='Thu nhập hàng năm (k$)')
    )
)

fig = go.Figure(data=[trace1], layout=layout)
py.iplot(fig)
```
